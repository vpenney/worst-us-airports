```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Read in airport coordinates spreadsheet
coordinates = pd.read_csv('airport_codes_mapped.csv')
coordinates.head()
```

```python
# Read in airline delays info
tsa_df = pd.read_csv('airline_delays.csv')
tsa_df.head()
```

```python
# Merge airline delays with coordinates
airports_full = coordinates.merge(tsa_df, left_on='locationID', right_on='airport')
```

```python
# Drop duplicate airport code column
airports_full.drop(columns=['locationID'], inplace=True)
```

```python
# Read in airport size
airports_size = pd.read_csv('2018_airport_size.csv')
airports_full = airports_full.merge(airports_size, left_on='airport', right_on='ORIGIN')
```

```python
departure_delays_2018 = pd.read_csv('2018_origin_delays.csv')
departure_delays_2018.sort_values(by='Avg(DEP_DELAY_NEW)', ascending=False).head()
```

```python
# Merge departure delays with coordinates for airports
departure_locations_2018 = coordinates.merge(departure_delays_2018, left_on='locationID', right_on='ORIGIN')
# Drop duplicate airport code column
departure_locations_2018.drop(columns=['locationID'], inplace=True)
# Rename minutes delayed column to something a bit snappier
departure_locations_2018.rename(columns={"Avg(DEP_DELAY_NEW)": "delay_minutes"}, inplace=True)
departure_locations_2018.sort_values(by='delay_minutes', ascending=False).head()
```

```python
departure_locations_2018 = departure_locations_2018.merge(airports_size, left_on='ORIGIN', right_on='ORIGIN')
departure_locations_2018.rename(columns={"Sum(FLIGHTS)": "sum_flights"}, inplace=True)
```

```python
# Export to csv
departure_locations_2018.to_csv('departure_locations_2018.csv')
```
