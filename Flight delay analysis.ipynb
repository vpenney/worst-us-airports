```python
import pandas as pd
```

## Read in all the 2018 pdfs, check dates

```python
# Categories: DayOfWeek, FlightDate, Reporting_Airline, Tail_Number
# OriginAirportID, Origin, OriginCityName, OriginState
# DestAirportID, Dest, DestCityName, DestState, DepTime, DepDelay
# ArrDelay, CarrierDelay, WeatherDelay, NASDelay, SecurityDelay, LateAircraftDelay
df1 = pd.read_csv('data/jan_2018.csv')
df2 = pd.read_csv('data/feb_2018.csv')
df3 = pd.read_csv('data/march_2018.csv')
df4 = pd.read_csv('data/april_2018.csv')
df5 = pd.read_csv('data/may_2018.csv')
df6 = pd.read_csv('data/june_2018.csv')
df7 = pd.read_csv('data/july_2018.csv')
df8 = pd.read_csv('data/aug_2018.csv')
df9 = pd.read_csv('data/sept_2018.csv')
```

```python
# df5.head()
```

## Join the separate months into a full dataframe for all of 2018

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], sort=True)
df
```

## Most Delayed Flights by origin and destination
### Grouped by the Origin and Destination airports, plus the average delay time for each route

```python
df.groupby(['ORIGIN','DEST'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=False)            
```

## Earliest departure flights, by origin and destination
### Grouped by origin and destination, with an average of the delay time

```python
df.groupby(['ORIGIN','DEST'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=True)
```

## Most delayed flights by origin, destination, and airline
### Grouped by Origin Airport, Destination, and Airline, with an average of the delay times

```python
df.groupby(['ORIGIN','DEST', 'OP_UNIQUE_CARRIER'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=False)
```

## Least delayed flights by origin, destination, and airline
### Grouped by Origin Airport, Destination, and Airline, with an average of the delay times

```python
df.groupby(['ORIGIN','DEST', 'OP_UNIQUE_CARRIER'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=True)
```

## Group by Origin, Destination, and Date to find potential bad weather days

```python
df.groupby(['ORIGIN','DEST', 'FL_DATE'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=False)
```

## Weather delays again, but this time using the WEATHER_DELAY column

```python
# Fill NaNs as zeros so the weather delay column averages properly
df.fillna(0).groupby(['ORIGIN','DEST', 'FL_DATE'], as_index=False)['WEATHER_DELAY'].mean().sort_values(by='WEATHER_DELAY', ascending=False)
```
