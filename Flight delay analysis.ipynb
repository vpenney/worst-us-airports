## Most Delayed Flights by origin and destination
### Grouped by the Origin and Destination airports, plus the average delay time for each route

```python
df18.groupby(['ORIGIN','DEST'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=False)            
```

## Earliest departure flights, by origin and destination
### Grouped by origin and destination, with an average of the delay time

```python
df18.groupby(['ORIGIN','DEST'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=True)
```

## Most delayed flights by origin, destination, and airline
### Grouped by Origin Airport, Destination, and Airline, with an average of the delay times

```python
df18.groupby(['ORIGIN','DEST', 'OP_UNIQUE_CARRIER'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=False)
```

## Least delayed flights by origin, destination, and airline
### Grouped by Origin Airport, Destination, and Airline, with an average of the delay times

```python
df18.groupby(['ORIGIN','DEST', 'OP_UNIQUE_CARRIER'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=True)
```

## Group by Origin, Destination, and Date to find potential bad weather days

```python
df18.groupby(['ORIGIN','DEST', 'FL_DATE'], as_index=False)['DEP_DELAY'].mean().sort_values(by='DEP_DELAY', ascending=False)
```

## Weather delays again, but this time using the WEATHER_DELAY column

```python
# Fill NaNs as zeros so the weather delay column averages properly
df18.fillna(0).groupby(['ORIGIN','DEST', 'FL_DATE'], as_index=False)['WEATHER_DELAY'].mean().sort_values(by='WEATHER_DELAY', ascending=False)
```
